```python
# 조건 : `참1`, `참2`, `거짓1` 이런 순으로 적어주세요.

# 개념시작
if ( True ):
  print("참")
    
if ( False ):
  print("거짓")
# 개념끝

# 문제시작
a = 10

# `==` => 같다.
if ( a == 10 ):
  print("참1")

# `!=` => 같지 않다.
if ( a != 10 ):
  print("거짓")

if ( a > 10 ):
  print("거짓2")

if ( a >= 10 ):
  print("참2")

b = 10

if ( a == b ):
  print("참3")

# c 에는 true or false 이 담긴다.
c = a != b

if ( c ):
  print("거짓3")

if ( c == False ):
  print("참4")

# `not` => 반전
if ( not c ):
  print("참5")

if ( not(not c) ):
  print("거짓4")

d = True

if ( c != d ):
  print("참6")

if ( 10 > 5 and -6 > -10 and 10 != 10 ):
  print("거짓5")

if ( 10 < 5 and -6 < -10 and 10 != 10 or 10 > 5 ):
  print("참7")
```

```python
# 문제 : 할인 대상인지 아닌지 출력해주세요.
# 조건 : 나이가 19세 이하이거나 60세 이상이면 할인 대상입니다.
# 조건 : `and, or`없이 풀어야 합니다.
# 조건 : 2가지 이상의 방법으로 풀어야 합니다.

age = 10

print("== 정답 v1 ==")
if(age<=19):
    if(age>60):       
        print("할인 대상입니다.")
if(age>19 or age<=60):
    print("할인 대상이 아닙니다.")

print("== 정답 v2 ==")
if(age<20 and age>60):  
    print("할인 대상입니다.")
else:
    print("할인 대상이 아닙니다.")
```

```python
# 문제 : 입력받은 정수가 소수인지 아닌지 알려주는 함수를 구현해주세요.

def is_prime_number(num):
  if num == 1:
    return False

  i = 2
  while i < num:
    if num % i == 0:
      return False
    i += 1
    
  return True

print("3은 소수입니다 : {}".format(is_prime_number(3)))
```

```python
'''
문제 : 딕셔너리에 각 달의 마지막 날들을 한번에 담기
조건 : 1월 ~ 3월까지
조건 : 2월의 마지막은 28일로 함
조건 : 첫 데이터의 key는 "1월", value는 31
'''
month_end_days = {}

for i in range(1, 12+1):
    month= str(i)+'월'
    if i==2:
        end_day=28
    elif i in[4,6,9,11]:
        end_day=30
    else:
        end_day=31
    month_end_days[month]=end_day

print(month_end_days)
```
